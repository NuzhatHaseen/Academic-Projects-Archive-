```python
#1. Copying the best model with a new chain id (To Superimpose)
!sed "s/ A / E /g" target.B99990011.pdb > bestmodel.pdb
#2. Setting up py3Dmol for visualization
view=py3Dmol.view()
#3. Loading template
view.addModel(open('4krr.pdb', 'r').read(),'pdb')
#4. Loading best DOPE score model
view.addModel(open('bestmodel.pdb', 'r').read(),'pdb')
#5. Zooming into all visualized structures
view.zoomTo()
#6. Here the background color was set as white
view.setBackgroundColor('white')
#7. Here the visualization style was set for chains
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
view.setStyle({'chain':'E'},{'cartoon': {'color':'yellow'}})
view.setStyle({'chain':'D'},{'cartoon': {'color':'green'}})
#8. Finally visualized the structures using the command below
view.show()
```
