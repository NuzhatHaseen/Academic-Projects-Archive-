# Getting 01_DEGsTreat.csv file

```r
library(dplyr);
library(limma);
library(edgeR);
```

```r
exprFinal = read.csv("00_ExpectedCnt.csv");
exprBT = exprFinal[, -c(1:3)];
rownames(exprBT) = exprFinal$gene;
exprBT = round(((2^exprBT)-1), 0);
write.csv(exprBT, "01_ExpectedCntBT.csv")
```

```r
expLIMMA = exprBT[, -which(names(exprBT) %in%
 c("GTEX.WFON.1426.SM.5CHT1",
"GTEX.SUCS.1026.SM.5CHTC",
"GTEX.WFG8.1726.SM.5CHTE",
"GTEX.WFG8.1526.SM.5CHSI"))]
```

```r
x = DGEList(expLIMMA)
```

```r
snames = colnames(x);
group = substr(snames, 1, 4); #Sets up level information for samples.
x$samples$group = group
```

```r
x = calcNormFactors(x, method = "upperquartile")
```

```r
design = model.matrix(~0 + group)
```

```r
colnames(design) = gsub("group", "", colnames(design));
contr.matrix = makeContrasts(TCGAvsGTEX = TCGA - GTEX,
 levels = colnames(design))
```

```r
v = voom(x, design, plot = TRUE)
```

```r
vfit = lmFit(v, design)
vfit = contrasts.fit(vfit, contrasts = contr.matrix)
```

```r
tfit = treat(vfit, lfc = 0.58);
```

```r
DEGsTreat = topTreat(tfit, n = Inf);
write.csv(DEGsTreat, "01_DEGsTreat.csv")
```

# Survival analysis with RegParallel

```r
library(dplyr);
library(data.table);
library(RegParallel);
library(survminer);
```

```r
dataExpr = read.csv("01_voomExpr.csv");
rownames(dataExpr) = dataExpr$X;

dataExpr = dataExpr %>% dplyr::select(starts_with("TCGA"));
head(dataExpr[1:4]);
```

```r
genes_toKeep = read.csv("02_GSandMM.csv");
head(genes_toKeep[1:5])
```

```r
genes_toKeep_list = genes_toKeep %>% filter(moduleColor == "darkred");
genes_toKeep_list = genes_toKeep_list$geneSymbol;

dataExpr = dataExpr[genes_toKeep_list, ];
head(dataExpr[1:4])
```

```r
dataExpr = as.data.frame(t(dataExpr));
dataExpr = as.data.frame(scale(dataExpr));
head(dataExpr[1:4]);
```

```r
highExpr = 0.674;
lowExpr = -0.674;
dataExpr = as.data.frame(ifelse(dataExpr <= lowExpr, 1,
 ifelse(dataExpr >= highExpr, 2, NA)));
dataExpr[] = lapply(dataExpr, factor);
dataExpr$SampleID = rownames(dataExpr)
```

```r
dataSurv = as.data.frame(fread("TCGA_survival_data"));
dataSurv$sample = gsub("-", ".", dataSurv$sample);
head(dataSurv);
```

```r
dataJoined = merge(dataSurv, dataExpr, by.x = "sample", by.y = "SampleID");
```

```r
temp4Hist = (dataJoined$OS.time)/365;
hist(temp4Hist,
 breaks = seq(from = 0, to = 15, by = 1),
 xaxt='n',
 xlab = "Years",
 main = "Histogram of OS.time");
axis(side = 1,
 at = seq(0, 15, 1),
 labels = seq(0, 15, 1));
```

```r
dataJoined = subset(dataJoined, OS.time <= 3650)
```

```r
res = RegParallel(data = dataJoined,
 formula = 'Surv(OS.time, OS) ~ [*]',
 FUN = function(formula, data) coxph(formula = formula,
 data = data,
 ties = 'breslow',
 singular.ok = TRUE),
 FUNtype = 'coxph',
 variables = colnames(dataJoined)[10:ncol(dataJoined)],
 blocksize = 10,
 cores = 2,
 nestedParallel = FALSE,
 conflevel = 95);
res = res[!is.na(res$P), ];
res = res[order(res$LogRank, decreasing = FALSE)];
head(res);
```

```r
write.csv(res, "04_Survival.csv")
```

# Cytoscape

```r
dataDEG = read.csv("01_DEGsTreat.csv");
head(dataDEG)
```

```r
dataCytoscape = merge(res, dataDEG, by.x = "Variable", by.y = "X");
dataCytoscape = dataCytoscape %>% dplyr::select(Variable, LogRank, logFC, adj.P.Val);
head(dataCytoscape);
```

```r
write.csv(dataCytoscape, "04_CytoscapeInput.csv")
```

# Short-listing genes and generating Km plot

```r
shortList = subset(dataCytoscape,
 LogRank < 0.1 & adj.P.Val < 0.1);
nrow(shortList)
```

```r
geneSL = shortList$Variable

for(i in 1:length(geneSL)) {
  fit = survfit(as.formula(paste0("Surv(OS.time, OS) ~", geneSL[i])), data = dataJoined)
  print(
    ggsurvplot(fit,
               pval = TRUE,
               risk.table = TRUE,
               break.time.by = 365,
               ggtheme = theme_bw(),
               palette = c("blue", "red"),
               xlim = c(0, 3650),
               risk.table.y.text.col = TRUE,
               risk.table.y.text = FALSE,
               tables.theme = theme_cleantable(),
               tables.height = 0.15,
               xlab = "Days",
               font.x = c(12, "bold"),
               ylab = "Survival Probability",
               font.y = c(12, "bold"),
               legend.labs = c("Low-expression", "High-expression"),
               font.legend = c(12, "bold"),
               legend.title = geneSL[i])
  )
}
```

# Differential correlation analysis with DGCA////////////////

```r
library(dplyr);
dataExpr = read.csv("01_voomExpr.csv");
head(dataExpr[1:6]);
```

```r
rownames(dataExpr) = dataExpr$X;
dataExpr = dataExpr[, -1];
head(dataExpr[1:6]);
```

```r
dataSample = names(dataExpr);
group = substr(dataSample, 1, 4);
dataGTEx = as.numeric(ifelse(group == "GTEX", 1, 0));
dataTCGA = as.numeric(ifelse(group == "TCGA", 1, 0));
dataMatrix = data.frame(dataGTEx, dataTCGA);
dataMatrix = as.matrix(dataMatrix);
```

```r
library(DGCA);
nrow(dataExpr);
```

```r
dataExpr = filterGenes(dataExpr,
 filterTypes = "central",
 keepRows = NULL,
 filterCentralType = "median",
 filterCentralPercentile = 0.25,
 allGroups = FALSE,
 design = NULL);
nrow(dataExpr);
```

```r
dataExpr = filterGenes(dataExpr,
 filterTypes = "dispersion",
 keepRows = NULL,
 filterDispersionType = "dispersion_index",
 filterDispersionPercentile = 0.25,
 allGroups = FALSE,
 design = NULL);
nrow(dataExpr);
```

```r
ddcor_DAPK3_PS = ddcorAll(inputMat = dataExpr,
 design = dataMatrix,
 inputMatB = NULL,
 compare = c("dataGTEx", "dataTCGA"),
 splitSet = "DAPK3",
 impute = FALSE,
 corrType = "pearson",
 nPairs = "all",
 sortBy = "zScoreDiff",
 adjust = "perm",
 nPerms = 1000,
 classify = TRUE,
 sigThresh = 0.05,
 corSigThresh = 0.05,
 heatmapPlot = FALSE,
 color_palette = NULL,
 verbose = FALSE,
 plotFdr = FALSE,
 corr_cutoff = 0.99,
 signType = "none",
 getDCorAvg = FALSE,
 oneSidedPVal = FALSE);
```
